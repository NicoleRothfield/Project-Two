```python
import sqlite3
conn = sqlite3.connect('my_data.db')
```

```python

import pandas as pd
# load the data into a Pandas DataFrame
users = pd.read_csv('Apr2021.csv')
# write the data to a sqlite table

users.to_sql('flightdata', conn, if_exists='append', index = False)
```

```python
query = "SELECT * FROM flightdata"

df = pd.read_sql(query, conn)

df.head()
```

```python
airline_agg = df.groupby(["MKT_UNIQUE_CARRIER"]).size().reset_index()

airline_agg.columns = ["Airline", "Count"]

airline_agg.head()
```
